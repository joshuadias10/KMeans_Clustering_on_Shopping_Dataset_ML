#**Implementation of K-Means Clustering using Shopping dataset - Elbow Method**

---



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the dataset
df = pd.read_csv("shoppingdata.csv")
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Display the last few rows of the dataset
print(df.tail())
```

```python
# Elbow method to determine the optimal number of clusters
distortions = []
K = range(1, 10)

for k in K:
    kmeans_model = KMeans(n_clusters=k, max_iter=25, random_state=1)
    kmeans_model.fit(df)
    distortions.append(kmeans_model.inertia_)
```

```python
# Plotting the Elbow Method graph
plt.figure(figsize=(16, 8))
plt.plot(K, distortions, 'bo-')
plt.xlabel("Number of Clusters")
plt.ylabel('Within-Cluster Sum of Squares (WSS)')
plt.title("Elbow Method for Optimal Number of Clusters")
plt.show()
```

```python
# Final model with the optimal number of clusters
optimal_clusters = 5  # Replace this with the optimal number from the Elbow plot
kmeans_final = KMeans(n_clusters=optimal_clusters, max_iter=25, random_state=1)
kmeans_final.fit(df)
```

```python
# Display the cluster centroids
print("Cluster Centers:\n", kmeans_final.cluster_centers_)
```

```python
# Display the cluster labels for each data point
print("Cluster Labels:\n", kmeans_final.labels_)
```

```python
# Scatter plot of clusters
plt.figure(figsize=(10, 6))
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=kmeans_final.labels_, cmap='viridis')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('K-Means Clustering of Shopping Data')
plt.show()
```
